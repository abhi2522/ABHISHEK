```python
# objct oriented method

# Create figure(empty canvas)
fig = plt.figure()

#adding set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width,height 

# plot on the set of axes
axes.plot(x,y,"b")
axes.set_xlabel("x_axis")
axes.set_ylabel("y_axis")
axes.set_title("title")
```

```python
# Create figure(empty canvas)
fig = plt.figure()

#adding set of axes to figure
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes # left,bottom,width,height 
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])  #inset axes

# plot on the main axes
axes1.plot(x,y,"b")
axes1.set_xlabel("x_axis")
axes1.set_ylabel("y_axis")
axes1.set_title("title1")


# plot on the main axes
axes2.plot(y,x,"r")
axes2.set_xlabel("x_axis")
axes2.set_ylabel("y_axis")
axes2.set_title("title2")

```

```python
fig = plt.figure() 

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend(loc=0) 
#loc=1 ---- upper right corner
#loc=2 ---- upper left corner
#loc=3 ---- lower left corner
#loc=4 ---- lower right corner
#loc=0 ---- optimal positon decided by python itself
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color="blue", alpha=0.5) #alpha is for transparency 
ax.plot(x, x+2, color="r")
ax.plot(x, x+3, color="#FFFF00")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color="blue", linewidth = 0.5,linestyle = "-", marker="+",markersize = 8)
ax.plot(x, x+2, color="r", lw=1.0, ls="-.", marker="o", markersize = 15) #o=cricle
ax.plot(x, x+3, color="#FFFF00",lw=2.0, ls=":", marker="s") #s=squre
```

```python
fig,axes = plt.subplots(1,2,figsize = (10,4))


axes[0].plot(x,x**2)
axes[0].plot(x,np.exp(x))
axes[0].set_title("nomal scale")
axes[0].grid(True)

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("logarithmic scale (y)")
axes[1].grid(color = "b",alpha = 0.5,linestyle = "dashed",lw = 0.5)
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#dist plot
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["tip"],kde = False,bins = 30)
```

### joint plot
* jointplot() allws you to matchup two distplot bivarite data with your choicei of what kind parameter you want to compare with.

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind = "scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind = "hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind = "reg")
```

```python
sns.pairplot(data=tips)
```
