```python
print("The minimum of budget is ",df["budget"].min())
print("The Q1 of budget is ",df["budget"].quantile(0.25)) #0.25=q1
print("The Mean of budget is ",df["budget"].mean())
print("The Median of budget is ",df["budget"].median()) #quantile(0.5)=q3
print("The Q3 of budget is ",df["budget"].quantile(0.75))
print("The Maxmimum of budget is ",df["budget"].max())
print("\n")
print("The Variance of budget is ",df["budget"].var())
print("The  Stdev of budget is ",df["budget"].std())
print("\n")
print("The Skewness of budget is ",df["budget"].skew())
print("The Kurtosis of budget is ",df["budget"].kurt())
```

```python
column = "revenue"
print(f"The minimum of {column} is ",df[column].min())
print(f"The Q1 of {column} is ",df[column].quantile(0.25)) #0.25=q1
print(f"The Mean of {column} is ",df[column].mean())
print(f"The Median of {column} is ",df[column].median()) #quantile(0.5)=q3
print(f"The Q3 of {column} is ",df[column].quantile(0.75))
print(f"The Maxmimum of {column} is ",df[column].max())
print("\n")
print(f"The Variance of {column} is ",df[column].var())
print(f"The  Stdev of {column} is ",df[column].std())
print(f"\n")
print(f"The Skewness of {column} is ",df[column].skew())
print(f"The Kurtosis of {column} is ",df[column].kurt())
```

```python
def des_stats(data,col):
    print(f"The minimum of {col} is ",data[col].min())
    print(f"The Q1 of {col} is ",data[col].quantile(0.25)) #0.25=q1
    print(f"The Mean of {col} is ",data[col].mean())
    print(f"The Median of {col} is ",data[col].median()) #quantile(0.5)=q3
    print(f"The Q3 of {col} is ",data[col].quantile(0.75))
    print(f"The Maxmimum of {col} is ",data[col].max())
    print("\n")
    print(f"The Variance of {col} is ",data[col].var())
    print(f"The  Stdev of {col} is ",data[col].std())
    print(f"\n")
    print(f"The Skewness of {col} is ",data[col].skew())
    print(f"The Kurtosis of {col} is ",data[col].kurt())
```

```python
des_stats(df,"budget")
```

```python
des_stats(df,"runtime")
```
