### WORKING WITH DEMOGRAPHIC DATA



```python
plt.figure(dpi=100)
print("total customer",df['customerID'].count())
data=df['gender'].value_counts()
print(data)

data.plot(kind='pie')
```

```python
#percentage of male and female  from total customer
p=df['customerID'].count()
male=((p-3488)/p)*100
female=100-male
print(f'the male customer are  {male}')
print(f'the female customers are  {female}')
```

```python
# number of seniorcitizen people with male female 
df1=df[['gender','SeniorCitizen']]
plt.figure(dpi=100)
ax=sns.countplot(data=df1,x='SeniorCitizen',hue='gender')
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.05, p.get_height()+0.05))
```

```python
# male and female and their partner info
plt.figure(dpi=100)
ax=sns.countplot(data=df,x='Partner',hue='gender')


for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.05, p.get_height()+0.05))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

```

```python
# infomation about partner and dependents
a=df.groupby(['Partner','Dependents']).count()['customerID']
a=pd.DataFrame(a)
print(a)


plt.figure(dpi=100)
ax=sns.countplot(data=df,x='Dependents',hue='Partner')


for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.05, p.get_height()+0.05))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

```

```python
print("various graph")
fig,axes=plt.subplots(1,2,figsize=(8,7),dpi=90)

axes[0].bar(df['Contract'],df['TotalCharges'],color="#DE952F")
axes[0].set_title('contract vs TotalCharges')
axes[0].set_xlabel("contract")
axes[0].set_ylabel("Totalcharges")



axes[1].bar(df['InternetService'],df['TotalCharges'],alpha=0.5)
axes[1].set_title('InternetService vsTotalCharges' )




```
