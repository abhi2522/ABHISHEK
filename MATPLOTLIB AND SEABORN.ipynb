```python
year = [2010,2012,2014,2016,2018,2020]
mango  = [0.91,0.89,0.95,0.88,0.89,0.988 ]
plt.plot(year,apple,"b*-",label = "apple")  

plt.plot(year,mango,"g+-",label = "mango")
plt.legend(loc = 0)
#add label
plt.xlabel("year")
plt.ylabel("ammount")

#add legend
plt.title ("crops growth")

 
```

```python
plt.plot(year,apple,marker = "o",label = "apple")  

plt.plot(year,mango,marker = "x",label = "mango")
plt.legend(loc = 0)
#add label
plt.xlabel("year")
plt.ylabel("ammount")

#add legend
plt.title ("crops growth")
```

* color/c
* linestype/ls
* linewidth/lw
* markersize/ms
* markeredgecolor/mec
* markeredgewidth/mew
* markeredgecolor/mcc
* alpha - opacity of plot
fmt = '[marker][line][color]'
            s    --     b
         


```python
plt.plot(year,apple,marker = "o",label = "apple",c="g",ls="-",lw=2,ms=8,mec="navy",mew=2)  

plt.plot(year,mango,marker = "x",label = "mango",c="r",ls="--",lw=2,ms=8,mew=3,alpha = 0.50)
plt.legend(loc = 0)
#add label
plt.xlabel("year")
plt.ylabel("ammount")

#add legend
plt.title ("crops growth")
 
```

```python
plt.plot(year,apple,"s-g",label = "apple",lw=2,ms=8,mec="navy",mew=2)  

plt.plot(year,mango,marker = "x",label = "mango",c="r",ls="--",lw=2,ms=8,mew=3,alpha = 0.50)
plt.legend(loc = 0)
#add label
plt.xlabel("year")
plt.ylabel("ammount")

#add legend
plt.title ("crops growth")
   
```
